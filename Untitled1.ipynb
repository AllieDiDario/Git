# Column Descriptions

Date - In format: yy-mm-dd

Open - The price of the stock at market open 

High - Highest price reached in the day

Low - Lowest price reached in the day

Close - The stock closing at the end of the Market hours

Adj Close - The closing price after adjustments for all applicable splits and dividend distributions.

Volume - Number of shares traded

```python
#only Date and High columns
df1=df[['Date','High']]
df1.head()
```

```python
#must rename columns ds and y for prophet to work
df1=df1.rename(columns={"Date" : "ds","High" : "y"})
df1.head()
```

```python
#define the model
a=Prophet(yearly_seasonality=True, weekly_seasonality=True, daily_seasonality=True,interval_width=0.95,)
a.fit(df1)
```

```python
#2 years
data=a.make_future_dataframe(periods=730)
```

```python
#last 5 rows
data.tail()
```

```python
#predict the labels of the data values
high=a.predict(data)
```

```python
#last 5 rows
high.tail()
```

```python
timeplot=high.iloc[-3000:,:]
plt.figure(figsize=[15,5])
plt.plot(timeplot['ds'],timeplot['yhat'],color='pink',linewidth=2,label='Highest Price')
plt.legend(loc='upper left')
plt.title('Highest price by Date')
plt.show()
```

```python
figure= a.plot(high)

for cp in a.changepoints:
    plt.axvline(cp,c='black',ls='--',lw=1)
```

```python
figure1=a.plot_components(high)
```

```python
highest=df['High']
components=df[['Open','Low','Close','Adj Close','Volume']]
x_train,x_test,y_train,y_test=train_test_split(
    components,highest,test_size=.25,random_state=0)
```

```python
lr=LinearRegression()
```

```python
x_train.shape
```

```python
#prediction
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
```

```python
print("mean squared error: {:.4f}".format(
    mean_squared_error(y_test,pred)))
print("variance score: {:.4f}".format(
    r2_score(y_test,pred)))
print("r-squared score: {:.4f}".format(
    lr.score(x_test,y_test)))
```
